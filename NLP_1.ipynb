```python
import string
string.punctuation
```

```python
def remove_punc(text):
    lol="".join([c.lower() for c in text  if c not in string.punctuation])
    return lol
```

```python
df['word1']=df['word'].apply(lambda l:remove_punc(l))
```

```python
import re 
def tokenize(lol):
    tokens=re.split("\W+",lol)
    return tokens
```

```python
df['token']=df['word1'].apply(lambda l: tokenize(l))
```

```python
import nltk
stp_words = nltk.corpus.stopwords.words('english')
len(stp_words)
```

```python
def remove_stop_words(lol1):
    lol1= [i for i in lol1 if str(i) not in stp_words]
    lol1
```
